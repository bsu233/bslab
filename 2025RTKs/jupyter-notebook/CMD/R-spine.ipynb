```python
import matplotlib.pyplot as plt
import numpy as np

# 加载数据
withjm_angle_rspine = np.loadtxt('/home/xhchen/VEGFR2/cmd-withJM_withoutJM/analysis/angle-Rspine/with-jm-angle_rspine.dat', comments="#")
withoutjm_angle_rspine = np.loadtxt('/home/xhchen/VEGFR2/cmd-withJM_withoutJM/analysis/angle-Rspine/without-jm-angle_rspine.dat', comments="#")

# 创建图形
plt.figure(figsize=(3, 2), dpi=200)

# 绘制直方图并获取直方图数据
n1, bins1, patches1 = plt.hist(withjm_angle_rspine[:,1], bins=30, alpha=0.4, 
                               label='with_jm', color='royalblue', density=True)
n2, bins2, patches2 = plt.hist(withoutjm_angle_rspine[:,1], bins=30, alpha=0.4, 
                               label='without_jm', color='red', density=True)

# 找出每个直方图中密度最大的区间
max_density_idx1 = np.argmax(n1)
max_density_idx2 = np.argmax(n2)

# 计算最大密度对应的角度值（使用区间的中心点）
max_angle1 = (bins1[max_density_idx1] + bins1[max_density_idx1+1]) / 2
max_angle2 = (bins2[max_density_idx2] + bins2[max_density_idx2+1]) / 2

# 添加垂直线标记最大密度点
plt.axvline(x=max_angle1, color='royalblue', linestyle='--', linewidth=1, alpha=0.8)
plt.axvline(x=max_angle2, color='red', linestyle='--', linewidth=1, alpha=0.8)

# 添加文本标注
plt.text(max_angle1+2, np.max(n1)*0.9, f'{max_angle1:.1f}°', 
         fontsize=5, color='royalblue', ha='left')
plt.text(max_angle2-2, np.max(n2)*0.9, f'{max_angle2:.1f}°', 
         fontsize=5, color='red', ha='right')

# 添加图例和标签
plt.legend(loc=1, fontsize=7, ncols=1)
plt.xlabel("Angle (\N{DEGREE SIGN})", fontsize=8)
plt.ylabel("density", fontsize=8)
plt.xticks(fontsize=8)
plt.yticks(fontsize=8)

# 保存图像
#plt.savefig('/home/xhchen/VEGFR2/cmd-withJM_withoutJM/analysis/angle-Rspine/rspine.png', 
 #           bbox_inches='tight', transparent=True)
plt.show()

# 打印峰值信息
print(f"With JM 峰值角度: {max_angle1:.2f}°")
print(f"Without JM 峰值角度: {max_angle2:.2f}°")
```
