```python
import numpy as np
import matplotlib.pylab as plt
from scipy.stats import gaussian_kde
from matplotlib.colors import ListedColormap,LinearSegmentedColormap
from matplotlib import cm
```

```python
with_jm_d1 = np.loadtxt('/home/xhchen/VEGFR2/cmd-withJM_withoutJM/analysis/dfg-conformation/withjm-d1.dat', comments="#")
with_jm_d2 = np.loadtxt('/home/xhchen/VEGFR2/cmd-withJM_withoutJM/analysis/dfg-conformation/withjm-d2.dat', comments="#")
```

```python
x = np.array(with_jm_d1[:, 1])
y = np.array(with_jm_d2[:, 1])
xy = np.vstack([x, y])
z = gaussian_kde(xy)(xy)

plt.figure(figsize=(3, 2), dpi=200)
plt.scatter(x, y, c=z, s=3, cmap='bwr', vmax=0.4, marker='.', alpha=0.8)

# 添加颜色条
scatter = plt.scatter(x, y, c=z, s=3, cmap='bwr', vmax=0.4, marker='.', alpha=0.8)
cbar = plt.colorbar(scatter, shrink=1.0)
cbar.set_label('Density', fontsize=8)
cbar.ax.tick_params(labelsize=7, length=2, width=0.8)  # 设置colorbar刻度线长度和宽度

plt.xlabel('d1 ($\AA$)', fontsize=8, labelpad=1)  # 调整x轴标签间距
plt.ylabel('d2 ($\AA$)', fontsize=8, labelpad=1)  # 调整y轴标签间距
plt.ylim([6, 15])
plt.xlim([3.5, 9])

# 添加浅绿色填充区域 - 框出 y >= 9 且 x <= 7.2 的区域
plt.fill_betweenx([9, 15], 3.5, 7.2, color='lightgreen', alpha=0.15)

# 添加L形虚线框
plt.axvline(x=7.2, ymin=(9-6)/(15-6), ymax=1, color='black', linestyle='--', linewidth=0.7, alpha=0.8)
plt.axhline(y=9, xmin=0, xmax=(7.2-3.5)/(9-3.5), color='black', linestyle='--', linewidth=0.7, alpha=0.8)

# 调整刻度线长度和样式
plt.tick_params(labelsize=7, length=2, width=0.5, pad=1)  # 调整刻度标签间距

# 调整布局
plt.tight_layout(pad=1)  # 调整整体边距
plt.savefig('/home/xhchen/VEGFR2/cmd-withJM_withoutJM/analysis/dfg-conformation/d1d2-with-jm.png',bbox_inches='tight',transparent=True)
#plt.savefig('/home/xhchen/VEGFR2/cmd-withJM_withoutJM/analysis/dfg-conformation/d1d2-with-jm.svg',bbox_inches='tight',transparent=True)

plt.show()
```
