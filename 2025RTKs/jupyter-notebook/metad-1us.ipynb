```python
import matplotlib.pyplot as plt
import numpy as np

# 
free_energy_1us = np.loadtxt('/home/xhchen/VEGFR2/metastable_jm/metaD_JM_metastable_sample/fes.dat', comments='#')
x = np.array(free_energy_1us[:,0] * 10)  # 
y = np.array(free_energy_1us[:,1] * 10)  # 
z = np.array(free_energy_1us[:,2] * 0.2389)  #  kcal/mol

# 
fig, ax = plt.subplots(figsize=(4,3), dpi=200)

#  - 
contourf = ax.tricontourf(x, y, z, cmap='rainbow', levels=20)

# 
contour_lines = ax.tricontour(x, y, z, colors='black', linewidths=0.5, levels=10, alpha=0.7)

# 
levels = contour_lines.levels

# -20-20
label_levels = [level for level in levels if level <= -20]

# 
ax.clabel(contour_lines, levels=label_levels, inline=True, fontsize=6, fmt='%1.1f')

# 
plt.xlabel('R1 ($\AA$)', fontsize=8, labelpad=2)  # labelpad
plt.ylabel('R2 ($\AA$)', fontsize=8, labelpad=2)  # labelpad

# 
ax.tick_params(axis='both', which='major', 
               labelsize=8, 
               length=3,          # 
               width=0.8,         # 
               pad=2)             # 

# 
cbar = plt.colorbar(contourf, label='Free Energy (kcal/mol)', orientation='vertical')
cbar.set_label('Free Energy (kcal/mol)', fontsize=8, labelpad=3)  # colorbar
cbar.ax.tick_params(labelsize=8, length=3, width=0.8, pad=2)  # colorbar

# 
plt.tight_layout()
plt.savefig('/home/xhchen/VEGFR2/metastable_jm/metaD_JM_metastable_sample/free_energy_1us-9-16.png',bbox_inches='tight',transparent=True)
plt.show()
```
