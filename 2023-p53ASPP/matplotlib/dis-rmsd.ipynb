```python
#ref-6rz3 p53-iaspp-rmsd-dis-heatmap
```

```python
dis=np.loadtxt('/home/teliu/work/aspp2/martini/p53/p53-iaspp/result/rmsd-dis/ref-6rz3-true/dis.dat',comments="#")
rmsd=np.loadtxt('/home/teliu/work/aspp2/martini/p53/p53-iaspp/result/rmsd-dis/ref-6rz3-true/ref-rmsd.dat',comments="#")
x=np.array(dis[:,1])
y=np.array(rmsd[:,1])
xy=np.vstack([x,y])
z=gaussian_kde(xy)(xy)
from matplotlib.colors import ListedColormap,LinearSegmentedColormap
from matplotlib import cm
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
cmap = LinearSegmentedColormap.from_list('mycmap',[(0, 'b'),(0.25,'cyan'),(0.5,'lightgreen'),(0.75, 'yellow'),(1, 'r')])
```

```python
fig, ax = plt.subplots(figsize=(0.5, 5))
fig.subplots_adjust(bottom=0)   # 设置子图到下边界的距离

cmap = cmap
norm = mpl.colors.Normalize(vmin=0, vmax=0.003)

fig.colorbar(mpl.cm.ScalarMappable(norm=norm, cmap=cmap),
             cax=ax, orientation='vertical')
#plt.savefig('/home/teliu/work/aspp2/image/xiu/hm-colorbar.png',bbox_inches='tight',transparent=True)
```

```python
#fig, ax = plt.subplots()
#plt.title('3CD_3D',fontsize=12)
plt.figure(figsize=(3,2),dpi=300)
fig=plt.scatter(x, y, c=z, s=5, cmap=cmap,vmax=0.003,marker='.',alpha=1)
fig=plt.scatter(38.36,3.26,s=7,color='black',marker='.')
fig=plt.scatter(33.05,20.44,s=7,color='black',marker='.')
plt.xlim([0,150])
plt.ylim([0,60])
#plt.ylim([9.5,11.5])
#plt.xlim([33,37])
plt.xlabel('Distance ($\AA$)',fontsize=10)
plt.ylabel('RMSD ($\AA$)',fontsize=10)
#plt.colorbar()
plt.title('p53-iASPP',fontsize=10)
plt.yticks(fontsize=10)
plt.xticks(fontsize=10)
#plt.savefig('/home/teliu/work/aspp2/figure/fig4/dis-rmsd-iaspp.png',bbox_inches='tight',transparent=True)
#plt.savefig('/home/teliu/work/aspp2/image/xiu/dis-rmsd-iaspp.png',bbox_inches='tight',transparent=True)
```
