```python
import numpy as np
import matplotlib.pylab as plt
```

```python
data0=np.loadtxt('/home/teliu/work/aspp2/amber/temp/idp/pro-domain/3/rmsd.dat',comments="#")
data1=np.loadtxt('/home/teliu/work/aspp2/amber/temp/idp/linker/rmsd.dat',comments="#")
```

```python
Time_Per_Frame = 0.2 # ns
plt.figure(figsize=(3,1),dpi=300)
plt.plot(data0[:,0]*Time_Per_Frame,data0[:,1],color='orange',linewidth=0.3, label='Pro')
plt.plot(data1[:,0]*Time_Per_Frame,data1[:,1],color='yellow',linewidth=0.3, label='Linker')
plt.ylim([0,50])
plt.xlim([0,500])
plt.legend(loc=1,fontsize=7)
plt.xlabel('Simulation Time(ns)',fontsize=10)
plt.ylabel('RMSD($\AA$)',fontsize=10)
plt.yticks(fontsize=10)
plt.xticks(fontsize=10)
#plt.savefig('/home/teliu/work/aspp2/image/xiu/idp-rmsd.png',bbox_inches='tight',transparent=True)
```
