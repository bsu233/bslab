```python
data1=np.loadtxt('/home/teliu/work/aspp2/martini/temp/new/pp53l-aspp2/result/idp-dis/pro.dat',comments="#")
data2=np.loadtxt('/home/teliu/work/aspp2/martini/temp/new/pp53l-aspp2/result/idp-dis/lin.dat',comments="#")
l1=np.loadtxt('/home/teliu/work/aspp2/martini/temp/new/pp53l-aspp2/result/idp-dis/list',comments="#")
```

```python
x1=np.mean(data1,axis=1)
x2=np.mean(data2,axis=1)
```

```python
std1 = np.std(data1,axis=1)
std2 = np.std(data2,axis=1)
```

```python
plt.figure(figsize=(3,1),dpi=300)
plt.plot(l1,x1,color='orange',linewidth=0.3,label='pro')
plt.fill_between(l1,x1-std1,x1+std1,color='orange',alpha=0.2,lw=2)
#plt.plot(l1,x2,color='yellow',linewidth=0.2,label='linker')
#plt.fill_between(l1,x2-std2,x2+std2,color='yellow',alpha=0.5,lw=1.5)
plt.ylim([0,100])
plt.legend(loc=1,fontsize=5)
plt.xlabel('Simulation Time(ns)',fontsize=10)
plt.ylabel('Distance($\AA$)',fontsize=10)
plt.yticks(fontsize=10)
plt.xticks(fontsize=10)
plt.savefig('/home/teliu/work/aspp2/image/xiu/pro-dis.png',bbox_inches='tight',transparent=True)
```
