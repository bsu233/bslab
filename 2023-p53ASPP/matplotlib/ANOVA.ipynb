```python
import pandas as pd
import numpy as np
 
from scipy import stats
from statsmodels.formula.api import ols
from statsmodels.stats.anova import anova_lm
 
#这是观测值
A1 = [-7.25,-2.81,-5.98,-4.42,-4.65,-25.52,-15.95,-12.03,-9.16,-17.89,-4.16,-16.37,-11.86]
A2 = [-2.99,-12.38,-4.46,-2.59,-27.89,-7.69,-4.67,-6.47,-5.67,-6.47,-25.64]
 
data = [A1, A2]
# 方差的齐性检验
w, p = stats.levene(*data)
if p < 0.05:
    print('方差齐性假设不成立')
 
 
# 成立之后， 就可以进行单因素方差分析
f, p = stats.f_oneway(*data)
print(f, p) 
```

```python
stats.levene(A1,A2)
```

```python
stats.ttest_ind(A1,A2,equal_var=True)
```
