```python
import matplotlib.pylab as plt
import numpy as np
from matplotlib.colors import ListedColormap, LinearSegmentedColormap
```

```python
WT=np.loadtxt('/home/teliu/work/aspp2/martini/p53/p53-iaspp/result/rep-frame/frame/pdb/data',usecols=[2])
```

```python
H_WT=WT.reshape ((20,20))
```

```python
vmax=30
```

```python
cmap1 = LinearSegmentedColormap.from_list('mycmap',[(0, 'w'),(10/vmax, 'aquamarine'),(20/vmax, 'cyan'),(30/vmax, 'b')])
```

```python
position =np.arange(0,20,1)
#position=(0,5,15,20)
print (position)
```

```python
x=position+1
y=position+1
plt.figure(figsize=(7,7),dpi=100)
plt.imshow(H_WT,cmap=cmap1,vmin=0,vmax=30,origin='lower',aspect='equal')
plt.title( "RMSD Matrix of iASPP",fontsize=25)
plt.xlabel("Conformations", fontsize=25)
plt.ylabel('Conformations', fontsize=25)
#ax.set_ylim(ax.get_ylim()[::-1])

plt.yticks(position,y)
plt.xticks(position,x)
cax=plt.colorbar().ax
#plt.xlim([1,20])
#plt.ylim([1,20])
#plt.savefig('iaspp-rmsd.png',bbox_inches='tight',transparent=True)
```
