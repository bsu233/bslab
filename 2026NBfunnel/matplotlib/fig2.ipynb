```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# 创建图形
plt.figure(figsize=(10*cm, 8*cm))
plt.tick_params(axis='both', which='major', width=0.5, length=2)
# 定义颜色和标签
colors = ['#1f77b4', '#2ca02c', '#d62728', '#9467bd', '#8c564b']
labels = ['Rosetta', 'IRAD', 'BLUUES', 'ZRank', 'FireDock']

# 为每组数据使用不同颜色和标记
datasets = [allrosetta, allirad, allbluues, allzrank, allfiredock]
markers = ['o', '^', 'D', 'v', '<']

# 绘制每组数据的回归图
for i, (data, color, label, marker) in enumerate(zip(datasets, colors, labels, markers)):
    sns.regplot(
        data=data, 
        x=data[:, 0], 
        y=data[:, 1], 
        scatter_kws={
            "s": 2, 
            "color": color,
            "marker": marker,
            "alpha":0.2,
        },
        line_kws={
            "color": color,
            "linewidth": 0.8
        },
        ci=None,
        label=label
    )

# 设置坐标轴标签和样式
plt.xlabel("Experimental $\Delta$G (kcal mol$^{-1}$)")
plt.ylabel("Calculated $\Delta$G (kcal mol$^{-1}$)")

# 设置刻度
#plt.xticks(fontsize=8)
#plt.yticks(fontsize=8)

# 设置坐标轴范围
plt.xlim([-16, -6])
plt.ylim([-200, 0])

# 添加网格线（与文献图片风格一致）
#plt.grid(True, linestyle='--', alpha=0.7)

# 添加图例
#plt.legend(loc=4, frameon=True, fontsize=5)



# 调整布局
plt.tight_layout()

# 显示图形
plt.savefig('/home/teliu/work/nanobody/image/draft/fig2/all.png',bbox_inches='tight',transparent=True)
```
