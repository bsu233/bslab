```python
allrosetta=np.loadtxt('/home/teliu/work/nanobody/score/result/new-exp-rosetta',comments="#")
plt.figure(figsize=(4,3),dpi=200)
plt.tick_params(axis='both', which='major', width=0.5, length=2)
# 读取数据
x = allrosetta[:,0]
y = allrosetta[:,1]
# 计算相关性
corr = np.corrcoef(x, y)[0, 1]
# 绘制散点图
sns.regplot(data=allrosetta,x=allrosetta[:,0], y=allrosetta[:,1], scatter_kws={"s": 3,"color":"gray"},color='black',ci=None)
# 加标题和坐标轴标签
plt.title("The Correlation of Experiment and Rosetta",fontsize=11)
plt.xlabel("Experiment Energy (kcal/mol)",fontsize=11)
plt.ylabel("Binding Free Energy (REU)",fontsize=11)
plt.xticks(fontsize=10)
plt.yticks(fontsize=10)
plt.xlim([-16,-7])
plt.ylim([-80,0])
# 加相关性数据
plt.text(0.45, 0.1, "Correlation coefficient: {:.4f}".format(corr), transform=plt.gca().transAxes,fontsize=8)
plt.text(0.6, 0.05, "$r^2$: {:.4f}".format(corr**2), transform=plt.gca().transAxes,fontsize=8)
# 显⽰图象
#plt.show()
#plt.savefig('/home/teliu/work/nanobody/image/energy/all/rosetta.png',bbox_inches='tight',transparent=True)
```
