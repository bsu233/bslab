```python
def concatenateData(data):
    """
    using the numpy vstack function to concatenate data
    """
    N = len(data)
    inidata = data[0]
    for i in range(1,N):
        #print (i)
        #t = np.reshape(data[i],((1,-1)))
        inidata = np.hstack((inidata,data[i]))
        
    return inidata
        
```

```python
Xgrid, Ygrid = np.mgrid[5:150:300j, 0:8:300j]
```

```python
# kernal density
def estimateTheDensity(cn,dis,X=Xgrid,Y=Ygrid):
    """
    using the guasian kernal to estimate the density of 
    sampled ligands on RNA
    """
    alldis = concatenateData(dis)
    allcn = concatenateData(cn)
    
    
    positions = np.vstack([X.ravel(), Y.ravel()])
    values = np.vstack([allcn, alldis])
    kernel = stats.gaussian_kde(values)
    Z = np.reshape(kernel(positions).T, X.shape)
    
    return Z, allcn, alldis
```

```python
def avgggg(data):
    avg = data[0].reshape((1,-1))
    N=len(data)
    for i in np.arange(1,N):
        t = data[i].reshape((1,-1))
        avg = np.vstack((avg,t))
        
    return np.mean(avg, axis=0)*0.24
        
    
```
