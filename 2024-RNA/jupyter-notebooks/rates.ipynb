```python
def runAvg(data,npot=5):
    """
    running average
    """
    t = np.asarray(data)
    npoints = len(data)
    #print (npoints)
    
    newdata=[]
    for i in range(npot+1,npoints-npot):
        #print(i)
        p=np.mean(t[i-npot:i+npot])
        newdata.append(p)
    return newdata
```

```python
def InterE(data,dt=10,stopN=3000):
    """
    Manually calculte the integral of the interaction
    """
    
    area=0.0
    
    maxN = data[stopN]
    for i in range(0,stopN,dt):
        avgE=(data[i]+data[i+dt])/2
        height= maxN-avgE
        area += height*dt
        
    return area
    
        
```

```python
def SpecialInterE(data,dt=10,stopN=4800,MaxN=100):
    """
    Manually calculte the integral of the interaction
    """
    
    area=0.0
    
    #maxN = data[stopN]
    for i in range(0,stopN,dt):
        avgE=(data[i]+data[i+dt])/2
        height= MaxN-avgE
        area += height*dt
        
    return area
    
```
