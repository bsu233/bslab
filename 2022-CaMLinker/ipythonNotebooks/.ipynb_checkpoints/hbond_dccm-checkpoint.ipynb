```python
def readHB(linker,Ca,Subdir="Hbond",Target="Notarget"):
    "read in the hbond"

    
    # the hbond information of each domain
    # NN - Ndomain, 
    # LL - Ndomain
    # CC - Cdomain
    hbinfo = {}
    
    datadir = os.path.join(RootDir,Subdir,Target)
    for i in ['NN','LL','CC']:
        file = os.path.join(datadir,linker,Ca,i+"avghb.dat") # individual hbond informatio
        file2 = os.path.join(datadir,linker,Ca,i+"nhb.dat") # total number of hbonds
        
        hbnumber = np.loadtxt(file2,comments="#")
        hbres = extracResidueInfo(file)


        hbinfo[i] = [hbnumber,hbres]
        
  
        
    return hbinfo
```

```python

def extracResidueInfo(file):
    """
    Read the specific hbond info
    
    """
    allinfo=np.array([[0.0,0.0,0.0]])
    with open(file) as f:
        aa = f.readlines()
    k = 0
    for i in aa:
        #print ("xx")
        if not i.startswith("#"):
            Frac = i.split()[4]
            residue = []
            for j in [0,1]:
                res = i.split()[j].split("@")[0]
                indx = i.split()[j].split("@")[0].find("_")
                numb = int(res[indx+1:])
                residue.append(numb)
            # only record the information for residues that are separate by >4 residues
            if abs(residue[0]-residue[1]) <= 4:
                continue
            else:
                residue.append(Frac)
                k += 1
                if k == 1:
                    #rint ("xx")
                    allinfo = np.asarray(residue)
                else:
                    allinfo = np.vstack((allinfo,residue))
            
    indces = np.flip(np.argsort(allinfo[:,2],axis=0))   
    tt = allinfo[indces]
    indcx, uniqs = np.unique(tt[:,0:2],axis=0,return_index=True)
    return tt[uniqs]
        

        
```

```python
def plotFig(data,domain="C",area=300):
    """
    Plotting Hbond Lifetime
    """
    plt.scatter(data[:,0].astype(int),data[:,1].astype(int),\
            s=area,c=data[:,2].astype(float),cmap=cmap,vmin=0.1,vmax=0.6,\
                marker="s",alpha=0.6,edgecolors='black',linewidths=0.1)
    if domain == "C":
        plt.xlim([80,150])
        plt.ylim([80,150])
    elif domain == "N":
        plt.xlim([0,70])
        plt.ylim([0,70])
    #plt.colorbar()
    
```
