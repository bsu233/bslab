```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.pylab as plt
import os
import pandas as pd
```

```python
x1=np.loadtxt('/home/jinp/project/NLRP3/PRS/noNEK7/frame81032/result.csv',comments="correlations")

```

```python
x2=np.loadtxt('/home/jinp/project/NLRP3/PRS/addNEK7/mainstate1/frame46887/result.csv',comments="correlations")

```

```python
x3=np.loadtxt('/home/jinp/project/NLRP3/PRS/addNEK7/mainstate2/frame86234/result.csv',comments="correlations")
```

```python
data1 = pd.Series(x1)
```

```python
data2 = pd.Series(x2)
```

```python
data3 = pd.Series(x3)
```

```python
plt.figure(figsize=(4,3),dpi=300)
plt.hist(data1,density = True,color='mediumblue',label = 'S1$_{withoutNEK7}$', rwidth=0.9,alpha = 0.8)
data1.plot(kind = 'kde',color='darkblue')

plt.hist(data2,density = True,color='mediumpurple',label = 'S1$_{withNEK7}$', rwidth=0.9,alpha = 0.8)

data2.plot(kind = 'kde',color='rebeccapurple')





plt.hist(data3,density = True,color='mediumslateblue',label = 'S2$_{withNEK7}$', rwidth=0.9,alpha = 0.8)


data3.plot(kind = 'kde',color='darkslateblue')










plt.xlabel("C$_{i}$",fontsize=10)
plt.ylabel("Probability density",fontsize=10)
plt.legend()
plt.title('to Active state') 
plt.savefig('PRS_prob.png',bbox_inches='tight',transparent=True)
```
